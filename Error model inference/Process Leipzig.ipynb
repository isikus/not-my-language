```python
import re
import os
import tarfile
import nltk

import pandas as pd

from nltk.tokenize import word_tokenize

nltk.download("punkt")
```

```python
def process_leipzig_file(archive_path, corpus_name):
    corpus_path = archive_path[:-7]
    tar = tarfile.open(archive_path)
    tar.extractall()
    res_d = {
        "sent_id": [None],
        "word": [],
        "whitespaces_follow": []
    }
    with open(os.path.join(corpus_path, corpus_path+"-sentences.txt"), "r", encoding="utf-8") as inc:
        sents = [s for s in inc.read().split("\n") if s]
        for s in sents:
            sid, sent = s.split("\t")
            sent_id = corpus_name + "_" + sid
            words = [w for w in re.split(r"( +)|(\w+)", sent) if w]
            for word in words:
                if re.search(r"^ +$", word) and sent_id == res_d["sent_id"][-1]:
                    res_d["whitespaces_follow"][-1] = len(word)
                else:
                    res_d["sent_id"].append(sent_id)
                    res_d["word"].append(word)
                    res_d["whitespaces_follow"].append(0)
    res_d["sent_id"] = res_d["sent_id"][1:]
    df = pd.DataFrame(res_d)
    return df
```

```python
df = process_leipzig_file("rus_newscrawl-public_2018_10K.tar.gz", "newscrawl")

df
```
