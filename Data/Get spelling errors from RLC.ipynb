```
spelling_errs = pd.read_excel("rlc_search_results.xlsx")

spelling_errs = spelling_errs.loc[spelling_errs["Тег"].notna()]
spelling_errs = spelling_errs.loc[spelling_errs["Ошибка"].notna()]
spelling_errs = spelling_errs.loc[spelling_errs["Исправление"].notna()]
spelling_errs = spelling_errs.loc[spelling_errs["Ошибка"] != ""]
spelling_errs = spelling_errs.loc[spelling_errs["Исправление"] != ""]
spelling_errs = spelling_errs.loc[~spelling_errs.duplicated('Ошибка')]

spelling_errs = spelling_errs.loc[spelling_errs["Тег"].str.contains(pat = 'Graph|Hyphen|Space|Ortho|Translit|Misspell', regex = True)]
```
